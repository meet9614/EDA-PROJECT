```python
#Loading the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
hh=pd.read_csv('/Users/meetishere/Desktop/Dataset/Superstore.csv',encoding='latin1')
```

```python
# 1)   Initial Data Exploration 
# Displaying first 5 rows
print(hh.head())
hh.drop(columns="Postal Code",inplace=True)  #Droping unnecessary cols
```

```python
#Checking values
print(hh["Category"].unique())
print(hh["Sub-Category"].unique())
```

```python
# Checking columns, data types, and missing values
print(hh.info())
```

```python
# showing total statistics for columns
print(hh.describe())
```

```python
#getting info about total rows and cols
hh.shape
```

```python
#2)    Data Cleaning
#Handling missing values
print(hh.isna().sum())
hh.fillna(0,inplace= True)
```

```python
#droping duplicvate values
print(hh.drop_duplicates())
```

```python
#transforming date into datetime
hh['Order Date']= pd.to_datetime(hh['Order Date'])
hh['Ship Date']= pd.to_datetime(hh['Ship Date'])
print(hh.head())
```

```python
#3)      Basic Analysis
# Getting top 10 most profitable orders
Top_10=hh.sort_values('Profit',ascending=False).reset_index().head(10)
print(Top_10)
```

```python
# Getting orders with Sales > $1000
s1000=hh['Order ID'][hh['Sales']>1000.0]
print(s1000)
```

```python
# Getting orders from a specific region (west)
West_reg=hh['Order ID'][hh['Region']=='West']
print(West_reg)
```

```python
# Total sales by region
sales_by_region = hh.groupby('Region')['Sales'].sum()
print(sales_by_region)
```

```python
# Average profit by category
Avg_profit_by_category= hh.groupby('Category')['Profit'].mean().reset_index()
print(Avg_profit_by_category)
```

```python
#4)      Advanced Analysis 
#1. Calculating Profit Margin (%)
hh['Profit Margin']=(hh['Profit']/hh['Sales'])*100
print(hh.head(2))

# Average Profit Margin
avg_margin = hh['Profit Margin'].mean()
print(f"Average Profit Margin: {avg_margin:.2f}%")
```

```python
#2. Categorizing profit margin (High/Low)
hh['Profit Category']=np.where(hh['Profit Margin']>10,'High','Low')
print(hh.head(5))
```

```python
#3. Monthly sales trend
hh['Month_year']= hh['Order Date'].dt.to_period('M') #extracting month and year from order date
Monthly_sale=hh.groupby('Month_year')['Sales'].sum().reset_index()
print(Monthly_sale)
```

```python
#5)   Visualizing monthly sales

# Plotting monthly sales trend

Monthly_sale.plot(x='Month_year',y='Sales',kind='line',figsize=(10,5))
plt.title('Monthly Sales Trend')
plt.show()
# Peaks in November (Holiday Season)
```

```python
 # Bar plot of sales by region
sales_by_region.plot(x='Region',y='Sales',kind='bar')
plt.title('Sales by region')
plt.show()
sales_by_region.plot.pie(autopct="%1.0f%%")
```

```python
# Bar plot of Profit by category
Avg_profit_by_category.plot(x='Category',y='Profit',kind='bar')
plt.title('Profit by Category')
plt.show()
```

```python
# Total Profit by Region

profit_by_region=hh.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print(profit_by_region)
profit_by_region.plot(x='Region',y='Profit',kind='bar')
plt.title('Profit by Region')
plt.show()
profit_by_region.plot.pie(autopct="%1.0f%%")
```

```python
#Sales by category

sales_by_Category = hh.groupby('Category')['Sales'].sum().reset_index().sort_values('Sales',ascending=False)
print(sales_by_region)

#Plotting
sales_by_Category.plot(x='Category',y='Sales',kind='bar')
plt.title('Sales by category')
plt.show()
```

```python
#Sales by state

sales_by_state=hh.groupby('State')['Sales'].sum()
print(sales_by_state)
sales_by_state.plot.bar()
```

```python
#Profit by States

Profit_by_state=hh.groupby('State')['Profit'].sum()
print(Profit_by_state)
Profit_by_state.plot.bar()
```

```python
#Customer Segmentation

#count orders per customer
customer_insight=hh['Customer ID'].value_counts().reset_index()
customer_insight.columns=['Customer ID','Order Count']

# Segment customers
customer_segment=np.where(customer_insight['Order Count']>1,'Regular Customer','One-tym Customer')
customer_insight['Customer segment']=customer_segment

# Analyze segment performance
segment_stats = hh.merge(customer_insight, on='Customer ID').groupby('Segment').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Customer ID': 'nunique'
}).rename(columns={'Customer ID': 'Customer Count'})

print("Customer Segmentation Analysis:\n", segment_stats)

#Plotting

segment_stats.reset_index().plot(x='Segment',y='Sales',kind='bar')
plt.title('Sales by Customer Segment')
plt.show()
```
